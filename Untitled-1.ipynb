2. Data Cleaning

○ Convert price-related columns (discounted_price, actual_price) to numerical format.

Part 2: MySQL Database Integration
4. Create a MySQL Database

○ Set up a MySQL database (ecommerce_db).
○ Create a table (products) with appropriate data types for each column.
○ Load the amazon.csv data into MySQL.

```python
import pandas as pd
import mysql.connector

df = pd.read_csv(r"C:\Users\PC\Desktop\AMAZON\amazon.csv")

df["discounted_price"] = df["discounted_price"].str.replace("₹", "").str.replace(",", "").astype(float)
df["actual_price"] = df["actual_price"].str.replace("₹", "").str.replace(",", "").astype(float)
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "").astype(float)
df["rating_count"] = df["rating_count"].astype(str).str.replace(",", "").replace("nan", "0").fillna("0").astype(int)

conn = mysql.connector.connect(
    host="localhost", user="root", password="Nakurugirls2003!", database="ecomm_db"
)
cursor = conn.cursor()

for _, row in df.iterrows():
    sql = """
    INSERT IGNORE INTO products (
        product_id, product_name, category, discounted_price, actual_price, discount_percentage, 
        discount_amount, rating, rating_count, about_product, img_link, product_link
    ) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    data = (
        row["product_id"],
        row["product_name"],
        row["category"],
        row["discounted_price"],
        row["actual_price"],
        row["discount_percentage"],
        row.get("discount_amount", 0),  
        row.get("rating", 0.0),  
        row.get("rating_count", 0),  
        row.get("about_product", ""),  
        row.get("img_link", ""),  
        row.get("product_link", "")  
    )

    cursor.execute(sql, data)  

conn.commit()
cursor.close()
conn.close()
print("Data inserted successfully!")
```

Part 3: Data Analysis in Python
6. Exploratory Data Analysis (EDA)

○ Show the distribution of ratings using a histogram.

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.histplot(df["rating"], bins=20, kde=True, color="blue")

# Customizing the plot
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Ratings", fontsize=14)
plt.grid(True)

plt.show()
```

○ Find the correlation between discount_percentage and rating.

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df = df.dropna(subset=["rating", "discount_percentage"])
correlation = df["discount_percentage"].corr(df["rating"])
print(f"Correlation between discount percentage and rating: {correlation:.4f}")
```

○ Identify top-selling product categories.

```python
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")  # Converting rating_count to numeric

top_categories = (
    df.groupby("category")["rating_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)  
)

print(top_categories)
```

○ Analyze the relationship between discount and rating_count.

```python
df["discount_percentage"] = pd.to_numeric(df["discount_percentage"], errors="coerce")
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")

correlation = df["discount_percentage"].corr(df["rating_count"])
print(f"Correlation between discount percentage and rating count: {correlation:.4f}")
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=df["discount_percentage"], y=df["rating_count"], scatter_kws={"alpha": 0.5})
plt.xlabel("Discount Percentage")
plt.ylabel("Rating Count")
plt.title("Relationship between Discount and Rating Count")
plt.show()
```

7. Data Visualization

○ Create a bar chart showing the top 10 highest-rated products

```python
# Selects the top 10 highest-rated products
top_rated_products = df.nlargest(10, "rating")

plt.figure(figsize=(10, 9))
sns.barplot(x=top_rated_products["rating"], y=top_rated_products["product_name"], palette="viridis")

plt.xlabel("Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Highest-Rated Products")
plt.xlim(0, 5)  
plt.xticks(rotation=45)
plt.show()
```

○ Plot a scatter plot of discount_percentage vs rating.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["discount_percentage"], y=df["rating"], alpha=0.5)

plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.title("Scatter Plot of Discount Percentage vs Rating")
plt.ylim(0, 5) 

plt.show()
```

○ Show the top-selling categories using a pie chart.

```python
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")  # Ensures rating_count is of numeric type
top_categories = df.groupby("category")["rating_count"].sum().sort_values(ascending=False).head(5) 

plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)

plt.title("Top-Selling Categories by Rating Count")
plt.show()
```

○ Use a heatmap to show correlations between numerical columns.

```python
# Computes the correlation matrix
corr_matrix = df[["discount_percentage", "rating", "rating_count", "discounted_price", "actual_price"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
